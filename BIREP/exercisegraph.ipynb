```python
import numpy as np
import matplotlib.pyplot as plt
nx,ny = 6,6
x = np.linspace(-3*np.pi/2, np.pi/2, nx)
y = np.linspace(-np.pi/2, 3*np.pi/2, ny)
xv,yv = np.meshgrid(x,y)

plt.xlim([-5,2.2])
plt.tick_params(left = False, right = False , labelleft = False ,
                labelbottom = False, bottom = False)
plt.plot(xv, yv, marker='o', color='k', linestyle='none')
for px in x:
  for py in y:
    plt.text(px+0.1,py+0.1, str(round(abs(np.cos(px)*np.cos(py))/16+0.01,2)))
    plt.plot([px,px],[py,y[-1]], color = 'k')
    plt.plot([px,x[-1]], [py,py], color = 'k')
for spine in plt.gca().spines.values():
    spine.set_visible(False)

plt.show()
```
